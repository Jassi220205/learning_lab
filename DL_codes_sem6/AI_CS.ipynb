```python
#heuristic
graph = {
    'Warehouse': ['Station A', 'Station B'],
    'Station A': ['Warehouse', 'Hub C', 'Hub D'],
    'Station B': ['Warehouse', 'Hub E'],
    'Hub C': ['Station A'],
    'Hub D': ['Station A', 'Goal'],
    'Hub E': ['Station B', 'Goal'],
    'Goal': ['Hub D', 'Hub E']
}
def dfs(graph, node, goal, visited=None):
    if visited is None:
        visited = set()
    visited.add(node)
    print(f"-> Visiting: {node}")
    if node == goal:
        return True
    for neighbor in graph.get(node, []):
        if neighbor not in visited:
            if dfs(graph, neighbor, goal, visited):
                return True
    return False
print("Available Locations:", ", ".join(graph.keys()))
start_node = input("Enter the starting location: ")
target_node = "Goal"
if start_node in graph:
    print(f"\n--- Starting DFS Search from {start_node} ---")
    if not dfs(graph, start_node, target_node):
        print("Goal not reachable from this location.")
    else:
        print("Success: Goal found!")
else:
    print("Invalid location entered.")
```

```python
#A*
import heapq
graph = {
    'Warehouse': [('Station A', 5), ('Station B', 10)],
    'Station A': [('Hub C', 3), ('Hub D', 8)],
    'Station B': [('Hub E', 2)],
    'Hub C': [('Goal', 12)],
    'Hub D': [('Goal', 2)],
    'Hub E': [('Goal', 15)],
    'Goal': []
}
heuristics = {
    'Warehouse': 15, 'Station A': 10, 'Station B': 12,
    'Hub C': 8, 'Hub D': 2, 'Hub E': 5, 'Goal': 0
}
def a_star(start, goal):
    pq = [(heuristics[start], start, [start], 0)]
    visited = {}
    while pq:
        f, current, path, g = heapq.heappop(pq)
        if current == goal:
            return path, g

        if current in visited and visited[current] <= g:
            continue
        visited[current] = g
        for neighbor, weight in graph.get(current, []):
            new_g = g + weight
            new_f = new_g + heuristics.get(neighbor, 100)
            heapq.heappush(pq, (new_f, neighbor, path + [neighbor], new_g))
    return None, None
print("Locations: Warehouse, Station A, Station B, Hub C, Hub D, Hub E, Goal")
user_start = input("Enter Start Location: ")
user_goal = input("Enter Destination: ")
if user_start in graph and user_goal in heuristics:
    path, total_cost = a_star(user_start, user_goal)
    if path:
        print(f"\n--- A* Navigation Results ---")
        print(f"Optimal Route: {' -> '.join(path)}")
        print(f"Total Distance/Cost: {total_cost}")
    else:
        print("No path found between those locations.")
else:
    print("Invalid locations. Please check your spelling.")
```
